```python
import numpy as np
import matplotlib.pyplot as plt
from torchvision import datasets

save_dir = "/Users/chensj/Documents/GitHub/paper_experiment_record/fedgra"
```

```python
# ── MNIST ────────────────────────────────────────────────────
mnist = datasets.MNIST(root="/tmp/data", train=True, download=True)

image0, _ = mnist[0]
image1, _ = mnist[1]
concat_image = np.concatenate([np.array(image0), np.array(image1)], axis=1)

plt.figure(figsize=(2, 1))
plt.imshow(concat_image, cmap="gray")
plt.axis("off")
plt.tight_layout(pad=0)
plt.savefig(f"{save_dir}/mnist_samples.pdf", bbox_inches="tight", pad_inches=0)
plt.show()
print("Saved: mnist_samples.pdf")
```
