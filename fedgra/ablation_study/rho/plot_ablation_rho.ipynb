# FedGRA rho ablation

Visualize the FMNIST and MNIST rho ablation results in one row.

```python
from pathlib import Path
import sys

cwd = Path.cwd()

if (cwd / "plot_ablation_rho.py").exists():
    # Running the notebook from fedgra/ablation_study/rho
    sys.path.insert(0, str(cwd))
    from plot_ablation_rho import plot_rho_ablation_mnist_fmnist
    base_dir = cwd
else:
    # Running the notebook from the repository root
    from fedgra.ablation_study.rho.plot_ablation_rho import plot_rho_ablation_mnist_fmnist
    base_dir = Path("fedgra/ablation_study/rho")
```

```python
fig, axes, runs = plot_rho_ablation_mnist_fmnist(
    base_dir=base_dir,
    datasets=("mnist", "fmnist"),  # left to right: MNIST, FMNIST
    metric="accuracy",
    smooth_window=5,
    plot_raw=True,
    figsize=(10, 5),
    legend_ncol=5,
    raw_linestyle="--",
    raw_zorder=1,
    line_zorder=3,
    legend_fontsize=14,
    legend_bbox_to_anchor=(0.5, 1.09),
    output_prefix=base_dir / "ablation_rho_mnist_fmnist",
    show=True,
)
```

```python
{dataset: list(rho_runs.keys()) for dataset, rho_runs in runs.items()}
```
